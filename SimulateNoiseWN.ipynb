# 1. Motivation: 
### we want to compare Weight Normaliztion (WN), Batch Normalization(BN) with the original setup, and want to see how much the improvement brought by the noise in WN than it does by batch normalization in term of trainig loss and validation accuracy.


## 2.  Experiment details:
We did experiment on dataset CIFAR-10 (50,000 trainning set, 10,000 test) and used the standard setup in LeNet (shown as below), i.e., 

- SGD with momuemtom =0.9 and batch-size = 128 images for training ( batch-size = 100 images for testing). 

- Learning rate lr = 0.1 at the first 70 or 80 epochs, lr = 0.01 between epochs (70 or 80) and (110 or 120) , lr = 0.001 after epochs 180 - 180, and lr = 0.0001 after epoch 180 or so (Whenever there is a large jump, it means we reset the learning rate)


## 3. Undertand the curve 

 - There are 4 plots (1st plot is about trainning loss, 2nd plot training accuracy, 3rd plot testing loss and 4th plot testing accuracy) 
 - X-axis is Epoch
 - Curvers with different colors means different setup. The legend on the to right or left means the following:
  - LeNet: standard setup as a baseline
  - LeNetBN: use batch normalization in all layers
  - LeNetWN: use weight normalization in all layers
  - LeNetWNnoise02: we inject Gaussian noise with mean 0 and standard deviation 0.2 to the training data.
  - LeNetWNnoise04mean_02: we inject Gaussian noise with mean -0.2 and standard deviation 0.4 at the training data.
  - LeNetWNnoise04toLayers: we inject Gaussian noise with mean 0 and standard deviation 0.2 at the training data as well as inputs of before the linear transform in 1st fully connected layer(after 2nd convalution layer) and inputs of before the linear transform in 2nd fully connected layer (after 1st FC layer)
